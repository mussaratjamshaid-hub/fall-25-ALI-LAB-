 Min-Max Algorithm

```python
import math


def run_minimax(level, pos, maxPlayer, values, lastLevel):
    if level == lastLevel:
        return values[pos]
    
    if maxPlayer:
        left  = run_minimax(level + 1, pos * 2, False, values, lastLevel)
        right = run_minimax(level + 1, pos * 2 + 1, False, values, lastLevel)

        return left if left > right else right
    else:
        left  = run_minimax(level + 1, pos * 2, True, values, lastLevel)

        right = run_minimax(level + 1, pos * 2 + 1, True, values, lastLevel)

        return left if left < right else right
    
leafValues = [3, 5, 2, 9, 3, 5, 2, 9]
treeDepth = int(math.log2(len(leafValues)))


answer = run_minimax(0, 0, True, leafValues, treeDepth)

print("Best possible outcome:", answer)
```
